```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import re
import nltk

from scipy import sparse
from sklearn.metrics.pairwise import pairwise_distances
from sklearn.preprocessing import StandardScaler

nltk.download('words')
```

```python
DATA_PATH = '/Users/olive/github/fantasynba/data/projections.xlsx'

nba = pd.read_excel(DATA_PATH)

nba[:10]
```

```python
nba.drop('TOTAL', axis=1, inplace=True)
nba.drop('R#', axis=1, inplace=True)
nba.drop('GP', axis=1, inplace=True)
nba.drop('TEAM', axis=1, inplace=True)
nba.drop('POS', axis=1, inplace=True)
```

```python
nba.isnull().any().any()
```

```python
words = set(nltk.corpus.words.words('en'))

def is_name(name):
    player_name = name.split()[:2]
    return ' '.join(player_name)
    
def remove_text_inside_brackets(text, brackets="()"):
    count = [0] * (len(brackets) // 2)
    saved_chars = []
    for character in text:
        for i, b in enumerate(brackets):
            if character == b: 
                kind, is_close = divmod(i, 2)
                count[kind] += (-1)**is_close 
                if count[kind] < 0: 
                    count[kind] = 0  
                else: 
                    break
        else:
            if not any(count):
                saved_chars.append(character)
    return float(''.join(saved_chars))
```

```python
nba['PLAYER'] = nba['PLAYER'].apply(is_name)

nba['FG%'] = nba['FG%'].apply(remove_text_inside_brackets)
nba['FT%'] = nba['FT%'].apply(remove_text_inside_brackets)

nba.head()
```

# Clustering players together

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

### Finding the optimal number of clusters

```python
scaler = StandardScaler()
X = nba._get_numeric_data().dropna(axis=1)
X_scaled = scaler.fit_transform(X)

pca_2 = PCA(2)
pca_2.fit(X_scaled)
X_pca = pca_2.transform(X_scaled)

kmeans_model = KMeans(n_clusters=20, random_state=1)
kmeans_model.fit(X_pca)
labels = kmeans_model.labels_

centers = kmeans_model.cluster_centers_
plt.scatter(x = X_pca[:,0], y = X_pca[:,1], c = labels)
plt.scatter(x = centers[:, 0], y = centers[:, 1], c = 'black', s = 200, alpha = 0.5)
```

```python
nba['Cluster'] = kmeans_model.labels_
nba[:12]

print(X_pca)
```

## DataFrame of separate clusters

### All around guards

```python
ALL_AROUND_GUARDS_20 = nba.loc[nba['Cluster'] == 1]
ALL_AROUND_GUARDS = ALL_AROUND_GUARDS_20.set_index(['PLAYER'])

ALL_AROUND_GUARDS 
```

```python
AG_sparse = sparse.csr_matrix(ALL_AROUND_GUARDS)
AG_sparse
```

### Recommendation System

```python
recommender = pairwise_distances(ALL_AROUND_GUARDS, metric='cosine')
```

```python
recommender_df = pd.DataFrame(recommender, index=ALL_AROUND_GUARDS.index, columns=ALL_AROUND_GUARDS.index)
```

```python
recommender_df.head()
```

```python
ALL_AROUND_GUARDS.mean()
```

```python
search = 'Stephen Curry'

for player in ALL_AROUND_GUARDS_20.loc[ALL_AROUND_GUARDS_20['PLAYER'].str.contains(search), 'PLAYER']:
    print('Players\n')
    print('5 closest players')
    print(recommender_df[player].sort_values()[1:6])
    print('')
```
